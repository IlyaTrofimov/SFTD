```python
import torch
import gudhi
import numpy as np
import matplotlib.pyplot as plt
from tqdm import trange
from sftd import plot_barcodes, SFTDLossGudhi
```

```python
G1 = torch.full((31, 31), 1)

G1[:, 0] = 0
G1[:, 10] = 0
G1[:, 20] = 0
G1[:, 30] = 0

G1[0, :] = 0
G1[10, :] = 0
G1[20, :] = 0
G1[30, :] = 0

G1[24:27, 0] = 1
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(3.5, 3.5))

plt.imshow(G1.numpy(), cmap = 'binary_r', origin = 'lower')
plt.colorbar(shrink = 0.8)
plt.show()
```

### Regular barcode

```python
cubical_complex = gudhi.CubicalComplex(vertices = G1.numpy())
cubical_complex.compute_persistence(homology_coeff_field = 2, min_persistence = 0.0)
barc0 = cubical_complex.persistence_intervals_in_dimension(0)[0:-1,]
barc1 = cubical_complex.persistence_intervals_in_dimension(1)
fig, ax = plt.subplots(1, 1)
fig.set_size_inches((3, 1))

plot_barcodes([barc0, barc1])
```

```python
fig, ax = plt.subplots(1, 1)
ax.set_ylim(-0.5, 4)
ax.set_xlim(-1.1, 0.75)
fig.set_size_inches((3, 1))

plot_barcodes(np.array([barc0, barc1]))
```

```python
G2 = torch.full((31, 31), 1)

G2[:, 0] = 0
G2[:, 10] = 0
G2[:, 20] = 0
G2[:, 30] = 0

G2[0, :] = 0
G2[10, :] = 0
G2[20, :] = 0
G2[30, :] = 0

G2[0, 24:27] = 1
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(3.5, 3.5))

plt.imshow(G2.numpy(), cmap = 'binary_r', origin = 'lower')
plt.colorbar(shrink = 0.8)
plt.show()
```

### Regular barcode

```python
cubical_complex = gudhi.CubicalComplex(vertices = G2.numpy())
cubical_complex.compute_persistence(homology_coeff_field = 2, min_persistence = 0.0)
barc0 = cubical_complex.persistence_intervals_in_dimension(0)[0:-1,]
barc1 = cubical_complex.persistence_intervals_in_dimension(1)
fig, ax = plt.subplots(1, 1)
fig.set_size_inches((3, 1))

plot_barcodes([barc0, barc1])
```

### SFTD Loss

```python
sftd_loss = SFTDLossGudhi(dims = [0, 1, 2])
```

```python
sftd_loss(G1, G2)
pers0 = sftd_loss.cubical_complex.persistence_intervals_in_dimension(0)
pers1 = sftd_loss.cubical_complex.persistence_intervals_in_dimension(1)
```

```python
fig, ax = plt.subplots(1, 1)
ax.set_ylim(-0.1, 1.2)
#ax.set_xlim(-1.1, 1.75)
fig.set_size_inches((3, 1))

plot_barcodes(np.array([pers0, pers1]))
```

```python
sftd_loss(G2, G1)
pers0 = sftd_loss.cubical_complex.persistence_intervals_in_dimension(0)
pers1 = sftd_loss.cubical_complex.persistence_intervals_in_dimension(1)
```

```python
fig, ax = plt.subplots(1, 1)
ax.set_ylim(-0.1, 1.2)
#ax.set_xlim(-1.1, 1.75)
fig.set_size_inches((3, 1))

plot_barcodes(np.array([pers0, pers1]))
```

```python
def get_elem(D_shape, num):
    i1 = num // (D_shape[0] * D_shape[1])
    num -= i1 * D_shape[0] * D_shape[1]
    i2 = num // D_shape[0]
    num -= i2 * D_shape[0]
    i3 = num

    return i3, i2, i1
```

```python
def add_markers(ax, dim = 1):
    for idx, (v1, v2) in enumerate(sftd_loss.v[0][dim]):
        print('dim', dim, v1, v2, get_elem(sftd_loss.D.shape, v1), get_elem(sftd_loss.D.shape, v2))

        markers = ['o', 'x', '*', 's', 'p']
        i1, i2, i3 = get_elem(sftd_loss.D.shape, v1)
        j1, j2, j3 = get_elem(sftd_loss.D.shape, v2)

        print('I', i1, i2, i3)
        ax.plot(i3, i2, 'o', color = 'red', markersize = 10)

        print('J', j1, j2, j3)
        ax.plot(j3, j2, 'o', color = 'orange', markersize = 10)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(3.5, 3.5))

ax.imshow(G1.numpy(), cmap = 'binary_r', origin = 'lower')

sftd_loss(G1, G2)
add_markers(ax)

sftd_loss(G2, G1)
add_markers(ax)
```
