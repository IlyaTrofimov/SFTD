```python
import os
import sys
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(-7.5, 10, 0.01)
```

```python
y = np.sin(x/1.75) + x**2 / 40
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

plt.plot(x, y)
plt.axline((-7.5, -0.835), (10.0, -0.835), linestyle = '--', color = 'r')
plt.text(-9.5, -0.835, '$y_1$', c = 'r', size = 15)

plt.axline((-7.5, 0.46), (10.0, 0.46), linestyle = '--', color = 'r')
plt.text(-9.5, 0.46, '$y_2$', c = 'r', size = 15)

plt.axline((-7.5, 1.22), (10.0, 1.22), linestyle = '--', color = 'r')
plt.text(-9.5, 1.22, '$y_3$', c = 'r', size = 15)
#plt.axis('off')

ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])

plt.show()
```
